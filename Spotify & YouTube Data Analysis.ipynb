```python
# removing the columns from the dataframe

data.drop(columns = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace = True)
```

## Q.1) Top 10 Artists - with the Highest Views on YouTube?

### Q.3) What are the most common Album Types on Spotify? How many tracks belong to each album type?

```python
z = data["Album_type"].unique()
z
```

```python
z_type = data.Album_type.value_counts()
z_type
```

```python
plt.pie(
    z_type,
    labels=z_type.index,
    autopct="%1.1f%%",
    startangle=60,
    colors=['m', 'y', 'r'],
    shadow=True,
    pctdistance=0.75
)
plt.show()
```

### Q.5) Top 5 YouTube Channels - based on the Views?

```python
data.head(1)
```

```python
yt = data.groupby('Channel')['Views'].sum().sort_values(ascending= False).head()
yt = yt.reset_index()
```

```python

sns.barplot( x = "Views", y = "Channel", data = yt, color='black')
plt.title('Top 5 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```
